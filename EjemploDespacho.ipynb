```python
import numpy as np
from scipy.optimize import minimize

# Función objetivo
def f(vars):
    Ps, Ph = vars
    return 1.15*(500+0.8*Ps+0.0016*Ps**2) #costo térmico

Pload=1200
Qmax=100e3/24
# Restricciones
cons = [
    # Igualdad: Ph+Ps-Pload-Ploss=0
    {'type': 'eq', 'fun': lambda v: v[0]+v[1]-8e-5*v[1]**2-Pload},

    # Desigualdad: 1500-Ps ≥ 0
    {'type': 'ineq', 'fun': lambda v: 1500-v[0]},

    # Desigualdad: Ps-150 ≥ 0
    {'type': 'ineq', 'fun': lambda v: v[0]-150},

    # Desigualdad: qmax-q ≥ 0
    {'type': 'ineq', 'fun': lambda v: Qmax-330-4.97*v[1]},
]

# Punto inicial
x0 = [0, 0]

resultado = minimize(f, x0, constraints=cons)

print(resultado.x)      # solución óptima
print(resultado.fun)    # valor mínimo de la función
```

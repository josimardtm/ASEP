```python
import cmath

def solve_quadratic(a, b, c):
  """Resuelve ecuaciones cuadráticas de la forma ax^2 + bx + c = 0.

  Args:
    a: Coeficiente de x^2.
    b: Coeficiente de x.
    c: Término constante.

  Returns:
    Una tupla con las dos soluciones de la ecuación. Si las soluciones son
    complejas, se devolverán como números complejos.
  """
  delta = (b**2) - 4*(a*c)

  if delta > 0:
    x1 = (-b - delta**0.5) / (2*a)
    x2 = (-b + delta**0.5) / (2*a)
  elif delta == 0:
    x1 = x2 = -b / (2*a)
  else:
    # Soluciones complejas
    x1 = (-b - cmath.sqrt(delta)) / (2 * a)
    x2 = (-b + cmath.sqrt(delta)) / (2 * a)
  return x1, x2
```

### Ejemplo de uso de la función `solve_quadratic`

Aquí te muestro cómo puedes usar la función para resolver una ecuación cuadrática. Por ejemplo, para la ecuación `x^2 - 5x + 6 = 0` (donde a=1, b=-5, c=6):

```python
# Ejemplo 1: Raíces reales distintas (x^2 - 5x + 6 = 0)
a1, b1, c1 = 1, -5, 6
sol1_x1, sol1_x2 = solve_quadratic(a1, b1, c1)
print(f"Las soluciones para x^2 - 5x + 6 = 0 son: {sol1_x1} y {sol1_x2}")

# Ejemplo 2: Raíces reales iguales (x^2 - 4x + 4 = 0)
a2, b2, c2 = 1, -4, 4
sol2_x1, sol2_x2 = solve_quadratic(a2, b2, c2)
print(f"Las soluciones para x^2 - 4x + 4 = 0 son: {sol2_x1} y {sol2_x2}")

# Ejemplo 3: Raíces complejas (x^2 + 2x + 5 = 0)
a3, b3, c3 = 1, 2, 5
sol3_x1, sol3_x2 = solve_quadratic(a3, b3, c3)
print(f"Las soluciones para x^2 + 2x + 5 = 0 son: {sol3_x1} y {sol3_x2}")
```
